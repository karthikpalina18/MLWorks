```python
df=pd.read_csv('/content/salaries.csv')
print("Name:Palina Karthik")
```

```python
df.head()
```

```python
df.columns
```

```python
input=df.drop('salary_more_then_100k',axis='columns')
```

```python
target=df['salary_more_then_100k']
```

```python
from sklearn import preprocessing
le_company=preprocessing.LabelEncoder()
le_job=preprocessing.LabelEncoder()
le_degree=preprocessing.LabelEncoder()
print("Name:Palina Karthik")
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_company.fit_transform(input['job'])
input['degree_n']=le_company.fit_transform(input['degree'])
```

```python
input.head()
```

```python
input_n=input.drop(['company','job','degree'],axis='columns')
```

```python
input_n.head()
```

```python
target.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
print("Name:Palina Karthik")
model.score(input_n,target)
```

```python
model.predict([[2,1,0]])
```

```python
model.predict([[2,1,1]])
```

```python
model.score(input_n,target)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
print("Name:Palina Karthik")
print(confusion_matrix(target,model.predict(input_n)))
print(classification_report(target,model.predict(input_n)))
print(accuracy_score(target,model.predict(input_n)))
```

```python
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt

model =DecisionTreeClassifier(criterion='entropy')
model.fit(input_n,target)
plt.figure(figsize=(15,10))
plot_tree(model,feature_names=df.columns[:-1])
plt.show()
```
