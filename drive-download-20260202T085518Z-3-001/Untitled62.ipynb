```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

# Load Wine dataset
wine = load_wine()
X = wine.data

# Convert to DataFrame for readability
df = pd.DataFrame(X, columns=wine.feature_names)

# Compute mean vector and covariance matrix
mean_vector = df.mean().values
cov_matrix = np.cov(df.T)

print("Mean Vector (first 5 features):\n", mean_vector[:5])
print("\nCovariance Matrix (shape):", cov_matrix.shape)

# Generate random samples from a multivariate normal distribution
np.random.seed(42)
num_samples = 500
samples = np.random.multivariate_normal(mean_vector, cov_matrix, num_samples)

# Convert samples to DataFrame
samples_df = pd.DataFrame(samples, columns=wine.feature_names)

# Show first few samples
print("\nGenerated Samples:")
print(samples_df.head())

# Plot a pair of features to visualize distribution
plt.figure(figsize=(7,5))
plt.scatter(samples_df["alcohol"], samples_df["malic_acid"], alpha=0.5, label="Generated")
plt.scatter(df["alcohol"], df["malic_acid"], alpha=0.5, label="Original")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.legend()
plt.title("Multivariate Normal Samples vs Original Data")
plt.show()
```
