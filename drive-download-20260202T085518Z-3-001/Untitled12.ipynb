```python
import statistics as st
```

```python
age =[20,19,20,18,21,22]
m_age=st.mean(age)
me_age=st.median(age)
mo_age=st.mode(age)
print("mean age :",m_age,"\nmedian age :",me_age,"\nmode age :",mo_age)
```

```python
age =[20,19,20,18,21,22,20]
va_age=st.variance(age)
sd_age=st.stdev(age)
print("variance age :",va_age,"\nstandard deviation age :",sd_age)
```

```python
age=[20,19,20,18,21,22,20]
r_age=max(age)-min(age)
print("range age :",r_age)
```
